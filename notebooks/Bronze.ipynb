# Source Dataset Collection

Three datasets are created for this project:

1. Agent Profile Data
2. Day 1 Customer Tickets
3. Day 2 Customer Tickets

These datasets simulate customer support operations and act as the input for the ETL pipeline.

````python
df_profiles = (
    spark.read
    .option("header", "true")
    .option("inferschema", "true")
    .csv(volume_path + "bronze/profiles.csv")
)
df_profiles.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(TBL_Profiles)

df_day1 = (
    spark.read
    .option("header", "true")
    .option("inferSchema","true")
    .csv(volume_path + "bronze/day1.csv")
)
df_day1.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(TBL_Day1)

df_day2 = (
    spark.read
    .option("header", "true")
    .option("inferSchema","true")
    .csv(volume_path + "bronze/day2.csv")
)
df_day2.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(TBL_Day2)
````
